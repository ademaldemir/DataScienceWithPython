```
#Filtering data in DataFrames

filt=mydf['Country']=='Turkey'
mydf[filt]
```

```
mydf.loc[filt]
mydf.loc[filt,'Age']
mydf.loc[filt, ['Age','ConvertedComp','DevType']]
```

```
#Specifying more conditions
# & : and
# | :  or 
#  ~ : not

filt= (mydf['Country']=='Turkey') & (mydf['DevType'].str.contains('Developer', na=False))
mydf.loc[filt]
```

```
filt= (mydf['Country']=='Germany') | (mydf['DevType'].str.contains('Developer', na=False))
mydf.loc[filt]
```

```
mydf.loc[~filt,['Country', 'DevType']] # show only country and devtype column values after filtering. 
```
